```python
def get_results(mname):
    run_and_save_res("grouped_religion", mname, task="SENT")
    run_and_save_res("grouped_sexuality", mname, task="SENT")
    run_and_save_res("grouped_gender", mname, task="SENT")
    run_and_save_res("grouped_disability", mname, task="SENT")
    run_and_save_res("grouped_age", mname, task="SENT")
    run_and_save_res("grouped_race", mname, task="SENT")
    run_and_save_res("grouped_country_by_gdp_ppp_quantile", mname, task="SENT")
    run_and_save_res("names_gender", mname, task="SENT")


def plot_class(mname, cl):
    core_name2test2res = {}
    for name, clean_name in names.items():
        core_name2test2res[clean_name] = load_results(name, mname)

    plot_all_tests_for_different_cores(
        model_name=mname,
        core_name2test2res=core_name2test2res,
        out_dir=f'{cfg.ROOT}/plots',
        cl=cl
    )

    
plot_class("roberta-sst-2", 1)

```
